```python
"""
Vamos a crear un programa que tenga el siguiente menú:

1. Añadir número a la lista: Me pide un número de la lista y lo añade al final de la lista.
2. Añadir número de la lista en una posición: Me pide un número y una posición, y si la posición existe en la lista lo añade a ella (la posición se pide a partir de 1).
3. Agregar un conjunto de datos a lista.
4. Eliminar el último número: Muestra el último número de la lista y lo borra.
5. Eliminar un número: Pide una posición, y si la posición existe en la lista lo borra de ella (la posición se pide a partir de 1).
6. Contar números: Te pide un número y te dice cuantas apariciones hay en la lista.
7. Posiciones de un número: recibe un número y te dice en que posiciones está (contando desde 1).
8. Mostrar lista: Muestra los números de la lista
9. buscar el mayor de la lista 
10. buscar el menor de la lista
11. ordenar por burbuja
S. Salir

"""

lista = []
def Agregar_pos(n,pos):
  if pos > len(lista):
    print("Posición incorrecta")
  else:
    lista.insert(pos - 1,n)
    
def Agregar():
  #llenas la lista
  fin = 's'
  while fin == 's':
    print('Ingrese un número:')
    n = int(input())
    lista.append(n)
    #valaidas las variable Fin
    while True:
      print('Desea ingresar otro número? Ingrese S para SÍ o N para NO:')
      fin = input()
      #convierto a minusculas
      fin=fin.lower()
      if fin == 's' or fin == 'n':
        break
      else:
        print('Carácter incorrecto, intente de nuevo.')

def eliminar_pos():
  pos = int(input("Dime una posición (empezando por 1):"))
  if pos > len(lista):
    print("Posición incorrecta")
  else:
    print("El elemento es %d y lo borramos." % lista.pop(pos - 1))

def buscar(n):
  print("el numero ",n," aparece ", lista.count(n))
  for pos,l in enumerate(lista):
    if l == n:
      print("está en la posición :",pos)

def mostrar():
  for num in lista:
    print(num," ",end="")
  print()

def buscar_M():
  mayor=lista[0]
  for n in lista:
    if n > mayor:
      mayor = n
  return mayor

def buscar_m():
  menor=lista[0]
  for n in lista:
    if n < menor:
      menor = n
  return menor

def burbuja():
  inter=True
  while inter:
    inter=False
    for i in range(len(lista)-1):
        if lista[i]>lista[i+1]:
          #intercambio los datos
          lista[i],lista[i+1]=lista[i+1],lista[i]
          inter=True
  mostrar()        

while True:
    print("\n")
    print("1. Añadir número a la lista")
    print("2. Añadir número de la lista en una posición")
    print("3. Agregar un conjunto de datos a lista.")
    print("4. Eliminar el último número")
    print("5. Eliminar un número por posición")
    print("6. Contar números")
    print("7. Posiciones de un número")
    print("8. Mostrar Lista")
    print("9. buscar el mayor de la lista")
    print("10. buscar el menor de la lista")
    print("11. ordenar por burbuja")
    print("12. Salir")
    opcion = int(input("Dime opción:"))
    if opcion == 1:
        num = int(input("Dime un número:"))
        lista.append(num)
    elif opcion == 2:
        num = int(input("Dime un número:"))
        pos = int(input("Dime una posición (empezando por 1):"))
        Agregar_pos(num,pos)
    elif opcion == 3:
        Agregar()
    elif opcion == 4:
        if len(lista)>0:
            print("El último elemento es %d y lo borramos." % lista.pop())
        else:
            print("La lista está vacía")
    elif opcion == 5:
         eliminar_pos()
    elif opcion == 6:
        num = int(input("Dime un número:"))
        print("El número %d aparece %d veces en la lista." % (num,lista.count(lista)))
    elif opcion == 7:
        num = int(input("Dime un número:"))
        buscar(num)
    elif opcion == 8:
        mostrar()
    elif opcion == 9:
        print("El mayor es: ", buscar_M())
    elif opcion == 10:
        print("El menor es: ", buscar_m())
    elif opcion == 11:
        burbuja()       
    elif opcion == 12:
        break
    else:
        print("Opción incorrecta")
```
